```python


def add_age_group(age):
    if type(age) == int:
        if age <= 25:
            return "jaunietis"
        if age > 25 and age < 62:
            return "pieaugusais"
        if age >= 62:
            return "seniors"
    else:
        return None

df.loc[:, "vecuma grupas"] = df['age'].apply(add_age_group)

df
```
