```python
prece1 = [28,143,55]
prece2 = [12, 45, 90]
prece3 = [56, 67, 89]

w1,w2,w3 = 0.3, 0.4, 0.3

weights = [w1,w2,w3]

def vertibu_aprekinasana(prece, weights):
    return prece[0] * weights[0] + prece[1] * weights[1] + prece[2] * weights[2]

vertibu_aprekinasana(prece1,weights)
```

```python
list(zip(prece1, weights))
```

```python
# v2

prece1 = [28,143,55]
prece2 = [12, 45, 90]
prece3 = [56, 67, 89]
w1,w2,w3 = 0.3, 0.4, 0.3
weights = [w1,w2,w3]

def vertibu_aprekins2(prece, weights):
    result = 0
    for p,w in zip(prece, weights):
        result += p*w
    return result

vertibu_aprekins2(prece1, weights)
```
